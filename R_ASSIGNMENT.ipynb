# **QUESTION 1 MONTECARLO**

```r
#MonteCarlo
```

```r
set.seed(123)
```

```r
# Number of simulations
num_simulations <- 1000
# Number of samples per interval
num_samples_per_interval <- 250
```

```r

# Exterior debris buildup (Normal Distribution)
exterior_mean <- 0.7
exterior_sd <- 0.4
exterior_debris <- matrix(rnorm(num_simulations * num_samples_per_interval, mean = exterior_mean, sd = exterior_sd), ncol = num_samples_per_interval)

# Interior debris buildup for the first engineer's opinion (Normal Distribution)
interior_mean_eng1 <- 0.5
interior_sd_eng1 <- 0.2
interior_debris_eng1 <- matrix(rnorm(num_simulations * num_samples_per_interval, mean = interior_mean_eng1, sd = interior_sd_eng1), ncol = num_samples_per_interval)

# Interior debris buildup for the second engineer's opinion (Uniform Distribution)
interior_min_eng2 <- 0
interior_max_eng2 <- 1
interior_debris_eng2 <- matrix(runif(num_simulations * num_samples_per_interval, min = interior_min_eng2, max = interior_max_eng2), ncol = num_samples_per_interval)

# Cumulative debris buildup for each simulation
cumulative_exterior_debris <- t(apply(exterior_debris, 1, cumsum))

cumulative_interior_debris_eng1 <- t(apply(interior_debris_eng1, 1, cumsum))
cumulative_interior_debris_eng2 <- t(apply(interior_debris_eng2, 1, cumsum))
#contamination
condition_matrix <- cumulative_exterior_debris > 180
#clogging
condition_matrix_cloging_eng1 = cumulative_interior_debris_eng1 > 130
condition_matrix_cloging_eng2 = cumulative_interior_debris_eng2 > 130
# # Distribution of interior and exterior buildup at the end of a service interval
exterior_at_service_end <- cumulative_exterior_debris[, ncol(cumulative_exterior_debris)]
interior_at_service_end_eng1 <- cumulative_interior_debris_eng1[, ncol(cumulative_interior_debris_eng1)]
interior_at_service_end_eng2 <- cumulative_interior_debris_eng2[, ncol(cumulative_interior_debris_eng2)]

# # Distribution of interior and exterior buildup before the end of a service interval
exterior_at_before_service_end <- cumulative_exterior_debris[, -ncol(cumulative_exterior_debris)]
interior_at_before_service_end_eng1 <- cumulative_interior_debris_eng1[, -ncol(cumulative_interior_debris_eng1)]
interior_at_before_service_end_eng2 <- cumulative_interior_debris_eng2[, -ncol(cumulative_interior_debris_eng2)]
# Visualize distributions
ggplot() +
  geom_density(aes(x = exterior_at_service_end, fill = "Exterior"), alpha = 0.5) +
  geom_density(aes(x = interior_at_service_end_eng1, fill = "Interior Eng 1"), alpha = 0.5) +
  geom_density(aes(x = interior_at_service_end_eng2, fill = "Interior Eng 2"), alpha = 0.5) +
  labs(title = "Distribution of Debris Buildup at Service End",
       x = "Debris Buildup (mg)",
       y = "Density",
       fill = "Component") +
  theme_minimal()

# Estimate likelihood of events
contamination_likelihood <- mean((exterior_at_before_service_end>180))


clogging_likelihood_eng1 <- mean(interior_at_before_service_end_eng1>130)
clogging_likelihood_eng2 <- mean(interior_at_before_service_end_eng2>130)

cat("Likelihood of contamination:", contamination_likelihood, "\n")
cat("Engineer 1 - Likelihood of clogging:", clogging_likelihood_eng1, "\n")

cat("Engineer 2 - Likelihood of clogging:", clogging_likelihood_eng2, "\n")


# Distribution of number of tests before events occur
tests_before_contamination <- apply(condition_matrix, 1, function(x) ifelse(any(x), min(which(x))-1, num_samples_per_interval))

tests_before_clogging_eng1 <- apply(condition_matrix_cloging_eng1> 130, 1, function(x) ifelse(any(x), min(which(x))-1, num_samples_per_interval))

tests_before_clogging_eng2 <- apply(condition_matrix_cloging_eng2  > 130, 1, function(x) ifelse(any(x), min(which(x))-1, num_samples_per_interval))

# Visualize distributions of tests before events
ggplot() +
  geom_density(aes(x = tests_before_contamination, fill = "Contamination Eng"), alpha = 0.5) +
  geom_density(aes(x = tests_before_clogging_eng1, fill = "Clogging Eng 1"), alpha = 0.5) +

  geom_density(aes(x = tests_before_clogging_eng2, fill = "Clogging Eng 2"), alpha = 0.5) +
  labs(title = "Distribution of Tests Before Events",
       x = "Number of Tests",
       y = "Density",
       fill = "Event") +
  theme_minimal()


```

# **QUESTION 2**

```r
# Define variables
cost_per_test <- 100
resample_cost <- 500
clog_cost <- 10000
storage_cost_per_sample <- 300
revenue_per_test <- 100

# Perform simulations for different numbers of specimens
num_simulations <- 1000
num_samples_per_interval <- 250
max_tests <- 250

profits <- numeric(max_tests)

for (num_tests in 1:max_tests) {
  total_cost <- 0
  total_revenue <- 0

  for (simulation in 1:num_simulations) {
    # Simulate contamination and clogging events
    contamination_event <- sample(c(TRUE, FALSE), num_tests, replace = TRUE)
    clogging_event <- sample(c(TRUE, FALSE), num_tests, replace = TRUE)

    # Calculate revenue and costs
    successful_tests <- sum(!contamination_event & !clogging_event)
    total_revenue <- total_revenue + successful_tests * revenue_per_test
    total_cost <- total_cost + sum(clogging_event) * (clog_cost+storage_cost_per_sample )+ sum(contamination_event) * resample_cost

  }

  # Calculate net profit
  profits[num_tests] <- total_revenue - total_cost
}

# Identify optimal strategy (number of specimens with maximum profit)
optimal_num_tests <- which.max(profits)
print(profits)
# Calculate expected net profit per day in the optimal scenario
expected_net_profit_per_day <- profits[optimal_num_tests] / num_simulations

# Print results
cat("Optimal Number of Tests:", optimal_num_tests, "\n")
cat("Expected Net Profit per Day (Optimal Scenario):", expected_net_profit_per_day, "\n")
```
